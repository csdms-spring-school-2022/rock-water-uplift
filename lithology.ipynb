# Exploring controls on long-term landscape evolution

## Import declarations

```python
import landlab
from landlab import RasterModelGrid
from landlab.components import FlowAccumulator, FastscapeEroder, LinearDiffuser, Lithology, LithoLayers, SinkFillerBarnes
from landlab.components.flow_accum.flow_accum_bw import find_drainage_area_and_discharge
from landlab.plot import imshow_grid
import numpy as np 
import time
import matplotlib.pyplot as plt
```

## Initialise Grid

We initialise a grid that is 100 km x 100 km wide, in units of *metres*. The boundaries are fixed at zero at the top and bottom but left closed at the east and west. Uniform noise up to 10 metres of elevation is added at all points in the landscape to allow for channelisation.

```python
nx = 100 # number of grid cells (x)
ny = 100 # number of grid cells (y)
dx = 1000 # metres

grid = RasterModelGrid((nx,ny),dx,xy_axis_units='m')
elev = grid.add_zeros('topographic__elevation',at='node')
grid.set_closed_boundaries_at_grid_edges(True,False,True,False)

np.random.seed(1)
elev[grid.core_nodes]+=np.random.uniform(0,10,size=elev.size)[grid.core_nodes]
imshow_grid(grid,'topographic__elevation',var_name="Elevation,m")
```

## Initialise Components

### Flowrouting

First we initialise the flowrouter and a sink-filler that fills in the pits in the random topography allowing for channelisation to occur. We display the drainage area that results from these calculations below.

```python
sb = SinkFillerBarnes(grid, method='D8')
sb.run_one_step()

fr = FlowAccumulator(grid, flow_director='D8')
fr.run_one_step()
imshow_grid(grid,'drainage_area',var_name="Drainage Area, m^2")
plt.xlabel("Easting, metres") 
plt.ylabel("North, metres")
plt.show()
```

### Erosion and diffusion

Next we initialise the stream power law eroder and the linear diffuser using realistic values. 

For the parameters $n$ and $m$ we use 1, and 0.5 respectively. 

For $K_{sp}$, a reasonable range appears to be $10^{-2}$ to $10^{-4}$ m/kyr. For $D_{diff}$, a reasonable range appears to be anything less than $10^{3}$ m$^2$/kyr. These ranges are broadly supported by real-life studies e.g. ([Roberts & White](https://agupubs.onlinelibrary.wiley.com/doi/10.1029/2009JB006692), [Rosenblum & Anderson](https://agupubs.onlinelibrary.wiley.com/doi/abs/10.1029/94JB00048)), although wider ranges have been reported (e.g. see [Campforts 2015](https://agupubs.onlinelibrary.wiley.com/doi/abs/10.1029/94JB00048))



```python
n_sp = 1
m_sp = 0.5 
K_sp = 1e-3 #m/kyr 
D_diff = 100 # m^2/kyr
diff = LinearDiffuser(grid,linear_diffusivity=D_diff)
sp = FastscapeEroder(grid, K_sp=K_sp,m_sp=m_sp,n_sp=n_sp)

peclet = ((nx*dx)*K_sp)/D_diff
print("Peclet number =",peclet)
```

#### Time-step calculations

Using these values we now calculate the appropriate timesteps so that these advective and diffusive timesteps can be performed correctly.

For the advective model, $dt < \frac{dx}{K_{sp}}$, and for the diffusive model, $dt < \frac{dx^2}{D_{diff}}$.

```python
dt_max_adv = grid.dx/K_sp
dt_max_diff = (grid.dx)**2/D_diff

print("Advective stability condition: t < ",dt_max_adv, "kyr")
print("Diffusive stability condition: t < ",dt_max_diff, "kyr")
```

Hence, as long as our timestep is less than these timesteps the solutions should be stable. 

We run the model, using a timestep of 100 kyr for 50 Myr at a uniform uplift rate of 0.1 m/kyr. 

```python
dt = 100 # 100 kyr
uplift_rate= 0.1 # 0.1 m/kyr
runtime= 50000 # 50000 kyr = 50 Myr
n_steps = runtime//dt 
print("Running for", n_steps, "iterations")

start=time.time()
for i in np.arange(n_steps):

    t=i*dt
    elev[grid.core_nodes] += uplift_rate * dt
    fr.run_one_step()
    sp.run_one_step(dt)
    diff.run_one_step(dt)
    
end=time.time()
print("Time per iteration =",round((end-start)/n_steps,5),"s")
print("Total runtime =",round((end-start),3),"s")
print("Peclet number =",peclet)
imshow_grid(grid,'topographic__elevation',var_name="Elevation, m",output=True,cmap='gist_earth')
imshow_grid(grid,'drainage_area',var_name="Drainage Area, m^2", output=True)
```

# Variable Lithology 

We wish to explore the impact of variable rock strength on erosion rates, for this we use the [`LithoLayers`](https://landlab.readthedocs.io/en/master/reference/components/litholayers.html) package. First, we renitiate our model grid so that it has a flat, zero elevation. 

Maximum amount of uplift (at 0.1m/kyr) for 50 Myr = 5000m.

```python
elev[:] = 0
random_start = np.random.uniform(0,10,size=elev.size)
elev[grid.core_nodes]+=random_start[grid.core_nodes]
sb.run_one_step()
fr.run_one_step()
```

## Horizontal layers - Increasing strength with depth

We first define 5 flat lithological layers (stored in `layer_elevations`). The _bottom_ boundaries of these layers lie at 1000, 2000, 3000, 4000, and 10000 metres below the surface. We only ever will erode 5000m of rock so the final layer we will never erode all the way through.

We assign each of these layers a different rock ID (`layer_ids`) so they can be distinguished. Next, we give each of these rock IDs a different erodibility ($K_{sp}$). The units get stronger with depth initially. Finally, we initialise the `LithoLayers` object. 

To make the varying lithologies impact on the erosion, we define a new `FastscapeEroder` that takes the $K_{sp}$ from the lithologies as input. Note, we don't include diffusion in this model as this relative simple mode for tracking lithologies does not consider lateral transport of material.

```python
layer_elevations = np.array([1000,2000,3000,4000,10000]) # Base of layers
layer_ids = np.array([0, 1, 2, 3, 4]) # Layer IDs
attrs = {'K_sp': {0: 1e-2, 1: 5e-3, 2: 1e-3, 3: 5e-4, 4: 1e-4}} # Erodibilities 
lith = LithoLayers(grid, layer_elevations, ids= layer_ids, attrs=attrs) # Initiate LithoLayers
sp_lith = FastscapeEroder(grid, K_sp='K_sp',m_sp=m_sp,n_sp=n_sp) # Link up erodibility to Fastscaper
```

The changing rock strength and ID with depth is shown below.

```python
id_arr = np.zeros(np.max(layer_elevations))
ksp_arr = np.zeros(np.max(layer_elevations))
for i in np.arange(layer_ids.size):
    elevation = layer_elevations[i]
    layer_id = layer_ids[i]
    ksp = attrs['K_sp'][layer_id]
    if(i==0):
        id_arr[0:elevation] = layer_id
        ksp_arr[0:elevation] = ksp
    else:
        id_arr[layer_elevations[i-1]:elevation] = layer_id
        ksp_arr[layer_elevations[i-1]:elevation] = ksp    
        
plt.figure(figsize=(4,4))
plt.subplot(1,2,2)        
plt.plot(ksp_arr,np.arange(np.max(layer_elevations)))
plt.xscale('log')
plt.xlabel("K_sp")
plt.ylim(0,5000)
plt.gca().invert_yaxis()

plt.subplot(1,2,1)
plt.plot(id_arr,np.arange(np.max(layer_elevations)))
plt.tight_layout()
plt.xlabel("Rock ID")
plt.ylabel("Depth beneath surface/m")
plt.ylim(0,5000)
plt.gca().invert_yaxis()
plt.show()
```

Now we have a system of uniformly eroded rocks. We progressively erode down through these units using the same initial model setup as above. Whenever we feed in uplift to our landscape we need to vertically _advect_ the lithologies upwards as well, this simulates progressive uplift of a stratigraphically layered unit.  

The chunk below outputs the topography, and the 'geological map' at 2.5 Myr timesteps. 

```python
dt = 100 # 100 kyr
uplift_rate= 0.1 # 0.1 m/kyr
runtime= 50000 # 50000 kyr = 50 Myr
n_steps = runtime//dt 
print("Running for", n_steps, "iterations")


plt.figure(figsize=(22,19))
counter = 0 # Counter for plotting
for i in np.arange(n_steps):
    t=i*dt
    fr.run_one_step() # Flow-route
    sp_lith.run_one_step(dt) # Erode the landscape one step
    uplift_step = uplift_rate * dt # Define the increase in elevation due to tectonic uplift
    dz_adv = uplift_step # Set the lithological advection to the tectonic uplift
    elev[grid.core_nodes] += uplift_step # Uplift the landscape
    lith.dz_advection=dz_adv # Set the advection of the litholayers object
    lith.run_one_step() # Advect the litholayers upwards
    
    if(t%2500==0):
        counter+=1 # Increment the plotting index
        plt.subplot(7,6,counter) # Add a subplot
        plt.imshow(grid.node_vector_to_raster(elev),cmap='gist_earth',vmin=0,vmax=1500,extent=(0,100000,0,100000),origin='lower') # Plot topo
        plt.title("t="+str(t/1000)+"Myr") 
        cb = plt.colorbar()
        cb.set_label("Elevation/m")
        counter+=1 # Increment the plotting index
        plt.subplot(7,6,counter) # Add a subplot    
        plt.imshow(grid.node_vector_to_raster(grid.at_node['rock_type__id']),cmap='magma',vmin=0,vmax=4,extent=(0,100000,0,100000),origin='lower') # Plot rock-type
        plt.title("Mean Ksp="+str(np.mean(grid.at_node['K_sp']))) # Display the average erodibility of the landscape
        cb = plt.colorbar()
        cb.set_label("Rock ID")
           
# Plot the final timestep            
print("t=",t/1000,"Myr")
counter+=1
plt.subplot(7,6,counter)
plt.imshow(grid.node_vector_to_raster(elev),cmap='gist_earth',vmin=0,vmax=1500,origin='lower')
cb = plt.colorbar()
cb.set_label("Elevation/m")
counter+=1
plt.subplot(7,6,counter)     
plt.imshow(grid.node_vector_to_raster(grid.at_node['rock_type__id']),cmap='magma',vmin=0,vmax=4,origin='lower')
plt.title("Mean Ksp="+str(np.mean(grid.at_node['K_sp'])))
cb = plt.colorbar()
cb.set_label("Rock ID")

plt.tight_layout()
plt.show()
```

Here we can see that as we uplift progressively older, more resistant rocks, the elevation of the steady state topography increases. Even under this constant uplift, the very low-resistance units at the surface are so quickly removed that no significant topography develops. Only when more resistant rocks are uplifted does significant topography develop.

## Tracking changing sediment provenance through time 

Now we track how this 'unroofing' scenario may be observed in sedimentary provenance records. To do this we define two functions that we will use to extract the proportion of sediment that is derived from a specific rock-type. 

These functions require an 'incision' amount to be defined. This is simply the difference between the pre-erosion timestep and the post-erosion timestep. To calculate the amount of a specific sediment that has been eroded we simply multiply this by the lithological map after the erosion timestep. Note that this assumes that the lithological change between timesteps is small, hence this is just an approximation. More accurate calculations (which e.g., take the average composition before and after the erosion) could easily be considered. 

```python
def get_sed_flux(rmg,dz):
    '''Returns the amount of sediment (m^3) being transported across landscape for 
    a given incision map. Divide by timestep length to return as a flux.
    This just wraps the `find_drainage_area_and_discharge` function.
    ____________
    rmg = RasterModelGrid object which has `flow__upstream_node_order` and 
    `flow__receiver_node` fields defined. This can be achieved by running a
    flow-routing component. 
    
    dz = array of size == rmg.size. This specifies the vertical change in surface
    elevation for an arbitary timestep. 
    '''
    a,tot_sed_flux = find_drainage_area_and_discharge(rmg.at_node['flow__upstream_node_order'], 
                                           rmg.at_node['flow__receiver_node'],
                                           node_cell_area = float(rmg.dx*rmg.dx),
                                           runoff = dz) 
    return(tot_sed_flux)

def get_sediment_comp(rmg,dz,rock_ID):
    '''Returns the proportion of the actively transported sediment that is derived
    from the provided rock_ID
    ____________
    rmg = RasterModelGrid object which has `flow__upstream_node_order` and 
    `flow__receiver_node` fields defined. This can be achieved by running a
    flow-routing component. 
    
    dz = array of size == rmg.size. This specifies the vertical change in surface
    elevation for an arbitary timestep. 
    
    rock_ID = float; This is the numeric ID which specifies a specific rock-type defined 
    in the rmg object.  
    '''

    tot_sed_flux = get_sed_flux(rmg,dz)
    outcrop_loc = rmg.at_node['rock_type__id']==rock_ID # Spatial map showing where particular rock type outcrops
    a, tot_rock_sed_flux = find_drainage_area_and_discharge(rmg.at_node['flow__upstream_node_order'], 
                                           rmg.at_node['flow__receiver_node'],
                                           node_cell_area = float(rmg.dx*rmg.dx),
                                           runoff = dz*outcrop_loc)
    with np.errstate(divide='ignore',invalid='ignore'): # Just to suppres div by zero warnings
        proportion=tot_rock_sed_flux/tot_sed_flux
    return(proportion)
```

With these functions define we now demonstrate their use for one timestep continuing from the landscape defined above. 

```python
fr.run_one_step()
elev_old = np.copy(elev) # Cache a copy of the elevation *before* erosion
sp_lith.run_one_step(dt) # Erode landscape one timestep
elev_new = np.copy(elev) # Cache a copy of the elevation *after* erosion
dz = elev_old-elev_new # Vertical change in elevation during timestep. 


sed_flux = get_sed_flux(grid,dz) # Get the sediment flux travelling through the system
prop_rock_3 = get_sediment_comp(grid,dz,3) # Get the proportion of sediment flux derived from Rock type 3
prop_rock_4 = get_sediment_comp(grid,dz,4) # Get the proportion of sediment flux derived from Rock type 4


# Plot the incision (dz)
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.imshow(dz.reshape(grid.shape),extent=(0,100000,0,100000),cmap='gray',origin='lower')
cb=plt.colorbar()
cb.set_label("dz, m")
         
# Plot the sediment flux
plt.subplot(2,2,2)
plt.imshow(sed_flux.reshape(grid.shape),extent=(0,100000,0,100000),origin='lower',cmap='hot')
cb=plt.colorbar()
cb.set_label("Sed volume, m^3")
         
# Plot the proportion of sediment flux derived from Rock type 3    
plt.subplot(2,2,3)
plt.imshow(prop_rock_3.reshape(grid.shape),extent=(0,100000,0,100000),cmap='magma',origin='lower')
cb=plt.colorbar()
cb.set_label("Prop. rock type 3")

# Plot the proportion of sediment flux derived from Rock type 4    
plt.subplot(2,2,4)
plt.imshow(prop_rock_4.reshape(grid.shape),extent=(0,100000,0,100000),cmap='magma',origin='lower')
cb=plt.colorbar()
cb.set_label("Prop. rock type 4")

plt.tight_layout()
plt.show()
```

### Generating a provenance record 

With these functions we can now see how the above erosional record may be recorded in the provenance record. To mimic a little more accurately how sediment is preserved, we track the average composition (and flux) of sediment that is leaving the bottom of the grid. 

We run the same landscape evolution scenario as above, but now at each time-step we calculate the composition of the sediment being eroded from the landscape and transported out of the southern boundary. We also track the *total* amount of sediment being eroded out too.  

```python
elev[:] = 0 # Reset elevation
np.random.seed(1)
random_start = np.random.uniform(0,10,size=elev.size)
elev[grid.core_nodes]+=random_start[grid.core_nodes]
sb.run_one_step()
fr.run_one_step()

layer_elevations = np.array([1000,2000,3000,4000,10000]) # Base of layers
layer_ids = np.array([0, 1, 2, 3, 4]) # Layer IDs
attrs = {'K_sp': {0: 1e-2, 1: 5e-3, 2: 1e-3, 3: 5e-4, 4: 1e-4}} # Erodibilities 
lith = LithoLayers(grid, layer_elevations,ids= layer_ids, attrs=attrs) # Initiate LithoLayers
sp_lith = FastscapeEroder(grid, K_sp='K_sp',m_sp=m_sp,n_sp=n_sp) # Link up erodibility to Fastscaper

dt = 100 # 100 kyr
uplift_rate= 0.1 # 0.1 m/kyr
runtime= 50000 # 50000 kyr = 50 Myr
n_steps = runtime//dt 
print("Running for", n_steps, "iterations")

sed_comp_record = np.zeros((n_steps,layer_ids.size)) # Initialise output sed. comp. array.
sed_flux_record = np.zeros((n_steps,1)) # Initialise output sed. flux. array.

for i in np.arange(n_steps):
    t=i*dt
    fr.run_one_step() # Flow routing
    elev_old = np.copy(elev) # Cache a copy of the elevation *before* erosion
    sp_lith.run_one_step(dt) # Erode one timestep
    elev_new = np.copy(elev) # Cache a copy of the elevation *after* erosion
    dz = elev_old-elev_new # Vertical change in elevation
    
    sed_flux_at_margin = get_sed_flux(grid,dz)[grid.y_of_node==0] # Extract the sediment flux along the southern margin
    tot_sed_at_margin = np.sum(sed_flux_at_margin) # Total flux coming out southern margin
    sed_flux_record[i] = tot_sed_at_margin # Update the out array of sediment flux 
    for rock_type in layer_ids: # Loop through each rock type ID
        sed_comp_at_margin = get_sediment_comp(grid,dz,rock_type)[grid.y_of_node==0] # Calculate sediment comp along southern margin
        sed_comp_record[i,rock_type] = np.nansum(sed_comp_at_margin*sed_flux_at_margin)/tot_sed_at_margin # Convert to *average* sediment comp at margin
    uplift_step = uplift_rate * dt 
    dz_adv = uplift_step # Advect lithological layers
    elev[grid.core_nodes] += uplift_step # Uplift topography
    lith.dz_advection=dz_adv # Advect lithological layers
    lith.run_one_step()
```

Now we visualise these synthetic 'provenance' records through time. 

```python
time=np.arange(0,runtime,dt)
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.title("Increasing strength with depth")
plt.plot(time/1000,sed_flux_record/dt)
plt.ylabel("Volumetric sed. flux, m3/kyr")

plt.subplot(2,1,2)
plt.plot(time/1000,sed_comp_record,label=[1,2,3,4,5])
plt.xlabel("Past <--- Time/Myr ---> Present")
plt.ylabel("Proportion of sediment for Rock type X")
plt.legend()
plt.show()
```

Here we can see the changing provenance of the output as well as the volumetric flux. As new lithologies are exposed, the landscape is put out of equilbrium and it takes some time for the new steady-state to be reached. For *more resistant rocks* the landscape takes longer to adjust. As, in this scenario, every new lithology is harder than the previous, the sediment flux traniently _decreases_. The amount of decrease is proportional to the relative change of $K_{sp}$. e.g., the sediment flux decrease from rock type 1 to 2 ($K_sp$ difference of 2x) is ~ half as big as the decrease from rock type 2 to 3 ($K_sp$ difference of 5x). If $K_{sp}$ decreases with new lithology, there is a transient sediment pulse as the new soft material is washed away.

```python
elev[:] = 0 # Reset elevation
np.random.seed(1)
random_start = np.random.uniform(0,10,size=elev.size)
elev[grid.core_nodes]+=random_start[grid.core_nodes]
sb.run_one_step()
fr.run_one_step()

layer_elevations = np.array([1000,2000,3000,4000,10000]) # Base of layers
layer_ids = np.array([0, 1, 2, 3, 4]) # Layer IDs
attrs = {'K_sp': {0: 5e-3, 1: 1e-3, 2: 5e-3, 3: 1e-3, 4: 5e-3}} # Erodibilities 
lith = LithoLayers(grid, layer_elevations, ids= layer_ids, attrs=attrs) # Initiate LithoLayers
sp_lith = FastscapeEroder(grid, K_sp='K_sp',m_sp=m_sp,n_sp=n_sp) # Link up erodibility to Fastscaper

dt = 100 # 100 kyr
uplift_rate= 0.1 # 0.1 m/kyr
runtime= 50000 # 50000 kyr = 50 Myr
n_steps = runtime//dt 
print("Running for", n_steps, "iterations")

sed_comp_record = np.zeros((n_steps,layer_ids.size)) # Initialise output sed. comp. array.
sed_flux_record = np.zeros((n_steps,1)) # Initialise output sed. flux. array.

for i in np.arange(n_steps):
    t=i*dt
    fr.run_one_step()
    elev_old = np.copy(elev)
    sp_lith.run_one_step(dt)
    elev_new = np.copy(elev)
    dz = elev_old-elev_new # Vertical change in elevation
    
    sed_flux_at_margin = get_sed_flux(grid,dz)[grid.y_of_node==0]
    tot_sed_at_margin = np.sum(sed_flux_at_margin)
    sed_flux_record[i] = tot_sed_at_margin
    for rock_type in layer_ids:
        sed_comp_at_margin = get_sediment_comp(grid,dz,rock_type)[grid.y_of_node==0]
        sed_comp_record[i,rock_type] = np.nansum(sed_comp_at_margin*sed_flux_at_margin)/tot_sed_at_margin
    uplift_step = uplift_rate * dt
    dz_adv = uplift_step
    elev[grid.core_nodes] += uplift_step    
    lith.dz_advection=dz_adv
    lith.run_one_step()

time=np.arange(0,runtime,dt)
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.title("Alternating hard/soft lithologies")
plt.plot(time/1000,sed_flux_record/dt)
plt.ylabel("Volumetric sed. flux, m3/kyr")

plt.subplot(2,1,2)
plt.plot(time/1000,sed_comp_record,label=[1,2,3,4,5])
plt.xlabel("Past <--- Time/Myr ---> Present")
plt.ylabel("Proportion of sediment for Rock type X")
plt.legend()
plt.show()
```

## Tilted strata of variable strengths

Here we explore the scenario of stratigraphic units dipping at ~8' to the NE. Each unit has a thickness of 2km but variable $K_{sp}$ between 0.005 and 0.001. Here we simply explore the impact that this has on the resulting drainage patterns. 

```python
elev[:] = 0 # Reset elevation
np.random.seed(1)
random_start = np.random.uniform(0,10,size=elev.size)
elev[grid.core_nodes]+=random_start[grid.core_nodes]
sb.run_one_step()
fr.run_one_step()

layer_elevations = np.arange(-20000,7000,2000)
layer_elevations[-1] += 1000
layer_ids = np.tile([0, 1], 7) 
attrs = {'K_sp': {0: 5e-3, 1: 1e-3}} # Erodibilities 
func = lambda x, y: ((0.1 * x) + (0.1 * y))

lith = LithoLayers(grid, layer_elevations,function=func, ids= layer_ids, attrs=attrs) # Initiate LithoLayers
sp_lith = FastscapeEroder(grid, K_sp='K_sp',m_sp=m_sp,n_sp=n_sp) # Link up erodibility to Fastscaper
imshow_grid(grid,'rock_type__id',var_name="Rock ID")
```

Having initiliased the grid and lithological column we now run the landscape evolution model.  

```python
dt = 100 # 100 kyr
uplift_rate= 0.1 # 0.1 m/kyr
runtime= 50000 # 50000 kyr = 50 Myr
n_steps = runtime//dt 
print("Running for", n_steps, "iterations")

plt.figure(figsize=(22,19))
counter = 0 # Counter for plotting
for i in np.arange(n_steps):
    t=i*dt
    fr.run_one_step() # Flow-route
    sp_lith.run_one_step(dt) # Erode the landscape one step
    uplift_step = uplift_rate * dt # Define the increase in elevation due to tectonic uplift
    dz_adv = uplift_step # Set the lithological advection to the tectonic uplift
    elev[grid.core_nodes] += uplift_step # Uplift the landscape
    lith.dz_advection=dz_adv # Set the advection of the litholayers object
    lith.run_one_step() # Advect the litholayers upwards
    
    if(t%2500==0):
        counter+=1 # Increment the plotting index
        plt.subplot(7,6,counter) # Add a subplot
        plt.imshow(grid.node_vector_to_raster(elev),cmap='gist_earth',vmin=0,vmax=500,extent=(0,100000,0,100000),origin='lower') # Plot topo
        plt.title("t="+str(t/1000)+"Myr") 
        cb = plt.colorbar()
        cb.set_label("Elevation/m")
        counter+=1 # Increment the plotting index
        plt.subplot(7,6,counter) # Add a subplot    
        plt.imshow(grid.node_vector_to_raster(grid.at_node['rock_type__id']),cmap='magma',vmin=0,vmax=4,extent=(0,100000,0,100000),origin='lower') # Plot rock-type
        plt.title("Mean Ksp="+str(np.mean(grid.at_node['K_sp']))) # Display the average erodibility of the landscape
        cb = plt.colorbar()
        cb.set_label("Rock ID")
           
# Plot the final timestep            
counter+=1
plt.subplot(7,6,counter)
plt.imshow(grid.node_vector_to_raster(elev),cmap='gist_earth',vmin=0,vmax=500,origin='lower')
cb = plt.colorbar()
cb.set_label("Elevation/m")
counter+=1
plt.subplot(7,6,counter)     
plt.imshow(grid.node_vector_to_raster(grid.at_node['rock_type__id']),cmap='magma',vmin=0,vmax=4,origin='lower')
plt.title("Mean Ksp="+str(np.mean(grid.at_node['K_sp'])))
cb = plt.colorbar()
cb.set_label("Rock ID")

plt.tight_layout()
plt.show()
```

Here we can clearly see that the variable strength imparts a fabric onto the resulting landscape and drainage pattern. As the rocks are eroded vertically down, the drainage evolves as the exposed rock types 'move' laterally to the NE. The weaker rock generates valleys and the more resistant rocks generate ridges.
